1.Salve a base de dados escolhida no seu drive e importe para um notebook do Colab, criando um Dataframe


```python
import pandas as pd
from google.colab import drive
drive.mount('/drive')
data=pd.read_csv(('/drive/My Drive/class_titanic.csv'))
```

2.Inspecione as colunas do conjunto de dados (Nome das colunas e tipo de dados)

```python
print(data.info())
```

3.Escolha uma coluna com valores numéricos e apague todas as linhas que possuam valor nulo

```python
nova_data=data.dropna(subset=['Age'])
display(nova_data)
print(nova_data.info())
```

4.Calcule a média desta mesma coluna

```python
media = nova_data['Age'].mean()
print(media)
```

5.Escolha um critério para eliminar as linhas que possuam valores muito distantes da média e crie um novo Dataframe.

```python
#critério: o valor tem que ser maior que 20 ou menor para ser distante



valor_dist1=nova_data['Age']<media+20
data3=nova_data[valor_dist1]
valor_dist2=nova_data['Age']<media-20
data3=data3[valor_dist2]
display(data3['Age'])





```

6.Crie um novo Dataframe que contenha apenas 5 colunas do conjunto anterior.

```python
data3.drop(columns=['Pclass','SibSp','Parch',	'Ticket',	'Fare',	'Cabin','Embarked'], inplace = True)

display(data3)
```

7.Escolha duas colunas numéricas e crie uma nova coluna que siga a regra: (Coluna A - Coluna B)*número aleatório (entre 0 e 1).

```python
import random
data3.insert(1, 'Coluna_nova',((data3['PassengerId']-data3['Age'])* random.random()).astype(int))
display(data3)
```

8.Embaralhe aleatoriamente o dataframe (índices e valores).

```python
import numpy as np

data4=data3.reindex(np.random.permutation(data3.index))

display(data4)
```

9.Ordene o Dataframe seguindo um critério definido por você (Explique o critério)

```python
#Ordenando o Dataframe de acordo com a ordem alfabetica da coluna Name

data4 = data4.sort_values(by='Name', ascending=True)
display(data4)
```

10.Escolha uma coluna numérica e um coluna com strings e utilize a função Groupby para mostrar os totais da  coluna.

```python
grupoNum = data4.groupby('PassengerId')[['PassengerId']].count()
grupoStr=data4.groupby('Name')[['Name']].count()

display(grupoNum)
display(grupoStr)
```

11.Utilize a função MAPS para fazer algum tipo de substituição no Dataframe (explique)

```python
#Substitui os números 1 e 0 da coluna 'Survived' pelas strings 'sobreviveu' e 'não sobreviveu'
dicionario={1:'Sobreviveu',0:'Não sobreviveu'}
data4['Sobreviveu']=data4['Survived'].map(dicionario)
display(data4['Sobreviveu'])


```
